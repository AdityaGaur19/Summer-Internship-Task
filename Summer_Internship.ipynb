# **Data Cleaning & Transformation**

```python
# Convert 'Month' to datetime for chronological processing
operational_costs['Month'] = pd.to_datetime(operational_costs['Month'], format='%B %Y')

# Remove leading/trailing spaces from 'Location' columns in all datasets
for df in [operational_costs, resource_utilization, incidents]:
    df['Location'] = df['Location'].str.strip()

# Convert numeric columns explicitly, coercing errors to NaN
resource_utilization['Utilization(%)'] = pd.to_numeric(resource_utilization['Utilization(%)'], errors='coerce')
resource_utilization['Downtime(hrs)'] = pd.to_numeric(resource_utilization['Downtime(hrs)'], errors='coerce')
incidents['ResponseTime(min)'] = pd.to_numeric(incidents['ResponseTime(min)'], errors='coerce')

# Convert incident 'DateTime' column to datetime type
incidents['DateTime'] = pd.to_datetime(incidents['DateTime'])

# Display counts of missing values in each dataset
print("Operational Costs Missing Values:\n", operational_costs.isnull().sum())
print("Resource Utilization Missing Values:\n", resource_utilization.isnull().sum())
print("Incidents Missing Values:\n", incidents.isnull().sum())
```

In this step, we begin by formatting the datasets to ensure data consistency and readiness for analysis. Converting the "Month" column from text to a datetime object in the operational costs dataset allows us to work efficiently with time series. Cleaning string fields by stripping excess spaces standardizes categorical data like warehouse locations, preventing grouping errors. Numeric columns such as utilization percentage, downtime hours, and incident response times are explicitly cast into numerical types, handling any invalid values gracefully via coercion. Converting incident timestamps into datetime objects facilitates precise temporal analyses. Finally, we check for any missing data, enabling informed decisions about handling null values moving forward.

# **Visualizing Monthly Operational Costs per Warehouse**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
ax = sns.lineplot(data=operational_costs, x='Month', y='OperationalCost(USD)', hue='Location', marker='o')

# Add labels for each data point for clarity
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=8)

plt.title("Monthly Operational Costs per Warehouse", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Cost (USD)")
plt.legend(title='Warehouse')
plt.tight_layout()
plt.show()
```

This visualization provides a clear, month-by-month comparison of operational costs across warehouses. The line plot helps to identify patterns such as cost increases, decreases, or anomalies, making it easy to detect trends over the six-month period. Including the numeric cost values directly on the plot enhances readability and allows for immediate interpretation without cross-referencing data tables.

# **Total Operational Cost Comparison by Warehouse**

```python
total_costs = operational_costs.groupby('Location')['OperationalCost(USD)'].sum().reset_index()

plt.figure(figsize=(8,5))
ax = sns.barplot(data=total_costs, x='Location', y='OperationalCost(USD)', palette='viridis')

# Annotate bars with precise total cost values
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=9)

plt.title("Total 6-Month Operational Costs by Warehouse", fontsize=16)
plt.xlabel("Warehouse")
plt.ylabel("Total Cost (USD)")
plt.tight_layout()
plt.show()
```

Aggregating operational costs across the timeline shifts focus from month-to-month details to a summarizing overview of total expenditure per warehouse. This bar chart clearly ranks warehouses by cumulative spending, equipped with value labels that provide concrete numbers for comparison. It offers a straightforward way to pinpoint which warehouses incur the highest costs, serving as a foundation for budget optimization discussions.

# **Resource Utilization Heatmap**

```python
util_pivot = resource_utilization.pivot(index='ResourceType', columns='Location', values='Utilization(%)')

plt.figure(figsize=(8,6))
sns.heatmap(util_pivot, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title("Resource Utilization (%) Across Warehouses")
plt.xlabel("Warehouse")
plt.ylabel("Resource Type")
plt.tight_layout()
plt.show()
```

The heatmap visualizes utilization rates of critical security resources, such as guards and CCTV cameras, across warehouses. This color-coded matrix quickly highlights which resources are highly utilized and which may be underused, providing actionable insights for balancing workload and reallocating assets. Annotated values make interpretation straightforward, ensuring stakeholders can easily grasp utilization differences at a glance.

# **Comparing Resource Downtime Across Warehouses and Types**

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(data=resource_utilization, x='ResourceType', y='Downtime(hrs)', hue='Location')

# Add numeric annotations above bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}', (p.get_x()+p.get_width()/2., p.get_height()),
                ha='center', va='bottom', fontsize=8)

plt.title("Resource Downtime by Type and Warehouse", fontsize=16)
plt.xlabel("Resource Type")
plt.ylabel("Downtime (hours)")
plt.legend(title='Warehouse')
plt.tight_layout()
plt.show()
```

Downtime impacts operational efficiency and security reliability. This grouped bar chart compares how long each resource type was unavailable in different warehouses, aiding the identification of vulnerabilities or maintenance issues. Showing exact downtime values fosters clear communication with maintenance teams and management, enhancing prioritization of repairs or replacements.

# **Incident Counts by Warehouse and Type**

```python
incident_counts = incidents.groupby(['Location', 'IncidentType']).size().reset_index(name='Count')

plt.figure(figsize=(12,6))
ax = sns.barplot(data=incident_counts, x='Location', y='Count', hue='IncidentType', dodge=True)

# Annotate with incident counts
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x()+p.get_width()/2., p.get_height()),
                ha='center', va='bottom', fontsize=8)

plt.title("Incident Counts by Type and Warehouse", fontsize=16)
plt.xlabel("Warehouse")
plt.ylabel("Number of Incidents")
plt.legend(title='Incident Type')
plt.tight_layout()
plt.show()
```

Breaking down the types and frequencies of safety incidents per warehouse makes it easier to identify warehouses with recurring issues or particular incident categories that dominate. This plotâ€™s grouped structure separates incident types, while the clear labels provide fast assessment of incident volumes, useful for targeted safety improvements.

# **Average Incident Response Time by Severity**

```python
avg_response = incidents.groupby('Severity')['ResponseTime(min)'].mean().reset_index()

plt.figure(figsize=(8,5))
ax = sns.barplot(data=avg_response, x='Severity', y='ResponseTime(min)', palette='mako')

# Adding response time labels on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}', (p.get_x()+p.get_width()/2., p.get_height()),
                ha='center', va='bottom', fontsize=8)

plt.title("Average Response Time by Incident Severity", fontsize=16)
plt.xlabel("Severity Level")
plt.ylabel("Average Response Time (minutes)")
plt.tight_layout()
plt.show()
```

Analyzing the average time taken to respond to incidents based on their severity serves as a vital performance metric. This bar chart allows assessment of whether high-severity incidents receive faster attention, which is crucial for minimizing damage or risk. Including numeric labels accentuates differences across severity levels, reinforcing areas needing process improvement.

# **Monthly Incident Trends per Warehouse**

```python
incidents['Month'] = incidents['DateTime'].dt.to_period('M').astype(str)
monthly_incidents = incidents.groupby(['Month', 'Location']).size().reset_index(name='Incidents')

plt.figure(figsize=(12,6))
ax = sns.lineplot(data=monthly_incidents, x='Month', y='Incidents', hue='Location', marker='o')

# Label each data point with incident counts
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=8)

plt.title("Monthly Incident Trends per Warehouse", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.tight_layout()
plt.show()
```

Narrative:
This time series plot tracks security incidents month-over-month for each warehouse. The visualization reveals temporal patterns, such as seasonal surges or drops in incidents, and helps in forecasting and resource allocation. The inclusion of data point labels improves the plotâ€™s communicative clarity, making it easier to report findings to stakeholders.

# **Correlation Between Operational Cost and Incident Count**

```python
inc_counts = incidents.groupby('Location').size().reset_index(name='NumIncidents')
cost_inc_merge = total_costs.merge(inc_counts, on='Location')

plt.figure(figsize=(8,5))
ax = sns.scatterplot(data=cost_inc_merge, x='OperationalCost(USD)', y='NumIncidents', hue='Location', s=100)

# Annotate points with warehouse names
for i, row in cost_inc_merge.iterrows():
    plt.text(row['OperationalCost(USD)']+40, row['NumIncidents'], row['Location'])

plt.title("Total Operational Cost vs Number of Incidents", fontsize=16)
plt.xlabel("Total 6-Month Operational Cost (USD)")
plt.ylabel("Incident Count")
plt.tight_layout()
plt.show()

# Print correlation coefficient
correlation = cost_inc_merge['OperationalCost(USD)'].corr(cost_inc_merge['NumIncidents'])
print(f"Correlation coefficient between cost and incidents: {correlation:.2f}")
```

Exploring the relationship between spending and incident occurrence is critical for evaluating the effectiveness of operational investments. The scatter plot presents each warehouseâ€™s total cost alongside incident counts, enriched with labels for clarity. Calculating the correlation coefficient quantifies this relationship numerically, providing an analytic basis for discussions on cost efficiency versus security outcomes.
